```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
players = pd.read_csv('data/player stats.csv')
players['Player'] = players['Player'].ffill()
players['Thrown Up'] = players['Thrown Up'].fillna(0)
players.head(3)
```

```python
positions = pd.read_csv('data/positions.csv')
players_merged = players.merge(positions, left_on='Player', right_on='Players')
players_merged = players_merged.drop(columns=['Players'])
players_merged
```

```python
players_merged.to_csv('data, player stats.csv', index=False)
```

```python
all_time = players_merged[['Player', 'Games', 'Goals', 'Assists','Clean Sheets','Thrown Up', 'Role', 'Position']].groupby('Player').agg({'Games':'sum', 'Goals':'sum','Assists':'sum','Clean Sheets':'sum', 'Thrown Up':'sum', 'Role':'max', 'Position':'max'})
all_time
```
